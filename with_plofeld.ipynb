```python
import plofeld
```

```python
from matplotlib import pyplot as plt
from plofeld.elements import PointCharge
from plofeld.fields import StaticField
from plofeld.utils.classes import Vector
from plofeld.utils.constants import ELECTRIC, MAGNETIC
```

```python
charges = [
    PointCharge(Vector(x=-1, y=1), q=1),
    PointCharge(Vector(x=1, y=1), q=-1),
    PointCharge(Vector(x=1, y=-1), q=1),
    PointCharge(Vector(x=-1, y=-1), q=-1),
]
```

```python
efield = StaticField(charges, field_type=ELECTRIC)
fe = efield.plot(xlim=(-2, 2), ylim=(-2, 2))
```

```python
bfield = StaticField(charges, field_type=MAGNETIC)
fb = bfield.plot(xlim=(-2, 2), ylim=(-2, 2))
```
