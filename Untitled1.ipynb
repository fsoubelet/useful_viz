```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve

%config InlineBackend.figure_format = "retina"
```

```python
x = np.linspace(-5.0, 5.0, 1000)
y = np.linspace(-5.0, 5.0, 1000)
X, Y = np.meshgrid(x, y)
Zn = X * Y
Zs = (X * X - Y * Y) / 2
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.set_aspect("equal")
ax2.set_aspect("equal")
ax1.set_title("Normal Quad")
ax2.set_title("Skew Quad")
ax1.set_xlabel("x")
ax2.set_xlabel("x")
ax1.set_ylabel("y")

cp_n = ax1.contour(X, Y, Zn, [-6, -4, -2, 2, 4, 6])
plt.colorbar()
cp_s = ax2.contour(X, Y, Zs, [-6, -4, -2, 2, 4, 6])
plt.colorbar()

ax1.quiver(X[::50, ::50], Y[::50, ::50], Y[::50, ::50], X[::50, ::50])
ax2.quiver(X[::50, ::50], Y[::50, ::50], X[::50, ::50], -Y[::50, ::50])
ax1.clabel(cp_n, inline=True, fontsize=3)
ax2.clabel(cp_s, inline=True, fontsize=3)
plt.colorbar(cp_n, cp_s)
plt.show()
```

```python
xpoints = np.arange(-5, 5, 0.05)
ypoints = np.arange(-5, 5, 0.05)
X, Y = np.meshgrid(xpoints, ypoints)

circularMask = False
areaRadius = 4
# order of the magnet
n = 2


def func(x, y):
    # the function to draw
    return ((x + 1j * y) ** (n)).real


func = np.vectorize(func)
V = func(X, Y)

# mask points which we don't want to draw
if circularMask:
    distance = np.sqrt(X**2 + Y**2)
    V = np.ma.masked_where(distance > areaRadius, V)
else:
    # polygonal mask
    # principal directions are at  (i + 0.5) / (2n) * 2pi
    for i in range(2 * n):
        angle = (i + 0.5) / float(2 * n) * 2 * np.pi

        # define a straight angle perpendicular to angle
        # mask all points on one side of this line
        anchor_x = areaRadius * np.cos(angle)
        anchor_y = areaRadius * np.sin(angle)

        normal_x = np.cos(angle)
        normal_y = np.sin(angle)

        def acceptFunc(x, y):
            value = (x - anchor_x) * normal_x + (y - anchor_y) * normal_y
            return value > 0

        acceptFunc = np.vectorize(acceptFunc)
        V = np.ma.masked_where(acceptFunc(X, Y), V)

if True:# levels equidistant in function value
    vmax = V.max()
    V /= vmax
    levels = np.arange(-2, 2, 0.05)
else:
    # levels equidistant on x and y axis
    # determine the levels to draw from values on one of the axes
    levels = [float(func(x, 0)) for x in arange(min(xpoints), max(xpoints), 0.50)] + [
        float(func(0, y)) for y in arange(min(ypoints), max(ypoints), 0.50)
    ]
    levels = sorted(list(set(levels)))
    vmax = 1

plt.figure(figsize=(6, 6))
Q = plt.contour(X, Y, V, colors="black", linestyles="solid", levels=levels)
# axis([-5,5,-5,5])
plt.xlabel("x coordinate")
plt.ylabel("y coordinate")

# mask points which we don't want to draw
if not circularMask:
    # polygonal mask
    # principal directions are at  (i + 0.5) / (2n) * 2pi
    for i in range(2 * n):
        angle = (i + 0.5) / float(2 * n) * 2 * np.pi
        if i % 2:
            label = "N"
            color = "red"
        else:
            label = "S"
            color = "green"
        anchor_x = 1.1 * areaRadius * np.cos(angle)
        anchor_y = 1.1 * areaRadius * np.sin(angle)
        plt.text(anchor_x, anchor_y, label, size=20, color=color, horizontalalignment="center", verticalalignment="center")

# ----------------------------------------
if n == 2:
    # quadrupole, draw some examples of force on charged particle
    # find kth level line on axes (x = 0 and y = 0)
    # the potential function is >= 0 on the x axis and <= 0 on the y axis
    # for a quadrupole
    lev = sorted(list(levels[levels >= 0]))[4]

    # find distance of this level on axis from origin
    # (exploit the 90 degree symmetry of the field)
    dist = fsolve(lambda x: func(x, 0) / vmax - lev, 3)[0]

    # rotation by +90 degrees
    rotMatrix = np.array([[0, -1], [1, 0]])
    invRotMatrix = rotMatrix.T

    arrowLength = 1.5
    arrowStart = np.array([dist, 0])
    origArrowDir = np.array([0, arrowLength])
    bfieldLabelPosOffset = np.array([0.3, 0.5 * arrowLength])
    forceLabelPosOffset = np.array([0.5 * arrowLength, -0.3])

    for i in range(4):
        arrowDir = origArrowDir[:]
        for j in range(i):
            arrowDir = rotMatrix.dot(arrowDir)

        # take into account quadrupole structure
        arrowDir *= (-1) ** i
        # draw arrow for the B field
        plt.arrow(arrowStart[0], arrowStart[1], arrowDir[0], arrowDir[1], head_width=0.3, head_length=0.3, color="red")

        # add a label for the B field
        textPos = arrowStart + (-1) ** i * bfieldLabelPosOffset
        plt.text(textPos[0], textPos[1], "B", size=20, color="red", horizontalalignment="center", verticalalignment="center")

        # draw the arrow for the force
        arrowDir2 = invRotMatrix.dot(arrowDir)
        plt.arrow(arrowStart[0], arrowStart[1], arrowDir2[0], arrowDir2[1], head_width=0.3, head_length=0.3, color="blue")

        # label for the force
        textPos = arrowStart + (-1) ** i * forceLabelPosOffset
        plt.text(textPos[0], textPos[1], "F", size=20, color="blue", horizontalalignment="center", verticalalignment="center")
        # ----------
        # prepare next iteration
        arrowStart = rotMatrix.dot(arrowStart)

        bfieldLabelPosOffset = rotMatrix.dot(bfieldLabelPosOffset)
        forceLabelPosOffset = rotMatrix.dot(forceLabelPosOffset)
```
